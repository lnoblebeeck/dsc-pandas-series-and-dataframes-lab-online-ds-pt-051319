## Rename all the columns to lower case

```python
#Your code here
new_head = [x.lower() for x in df.columns]
new_head
```

```python
df.columns = new_head
df.head()
```

## Change the Index to be the Line Names

```python
#Your code here
df = df.set_index('linename')
df.head()
```

## Painstakingly change the index back

```python
# Your code here
df = df.reset_index()
df.head()
```

## Create another column 'Num_Lines' that is a count of how many lines pass through a station. Then sort your dataframe by this column in descending order

```python
# Your code here
df['Num_lines'] = df.linename.map(lambda x: len(x))
df.head()
# df.info()
```

```python
df.sort_values(by=['Num_lines'],ascending=False, inplace = True)
# df.set_index('Num_lines', inplace = True)
```

## Compare subway traffic by day of the week. Display this as a graph.

```python
#Your code here
df.date = pd.to_datetime(df.date)
# print(df.date.dtype)

df['day_of_week'] = df.date.dt.dayofweek
df.head()
df.drop(columns=['Total Traffic', 'Total_Traffic'], inplace = True)
# df['Total_Traffic'] = df.entries + df.exits
# df.groupby('day_of_week')['Total_Traffic'].sum()
```

```python
df.groupby('day_of_week').sum().plot(kind = "line")
```

## Is there more subway traffic on a weekend or a weekday?    Be specific in comparing magnitudes

```python
#Your code here
df['Tota_traffic'] = df.entries + df.exits

df['Time_of_week'] = ''

df.loc[df['day_of_week'] < 5, "Time_of_week"] = "Weekday"
df.loc[df['day_of_week'] >= 5, "Time_of_week"] = "Weekend"
df.head()

df.groupby('Time_of_week')['Tota_traffic'].mean().plot("bar")
```
